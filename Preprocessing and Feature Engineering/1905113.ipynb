# Data cleaning

```python
df = df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'])
```

```python
# Remove/Replace missing values
df.dropna(subset=["Attrition"],inplace=True)
df.fillna(df.mode(), inplace=True)
df.shape
```

```python
# Remove duplicate values
df = df.drop_duplicates()
df.shape
```

# Creation of input and output features

```python
labels = df["Attrition"]
features = df.drop(["Attrition"], axis=1)
```

```python
labels.shape, features.shape
```

# Encoding

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoded_labels_np = encoder.fit_transform(labels)
encoded_labels_np
```

```python
# categorical_columns = ["Education", "EnvironmentSatisfaction", "JobInvolvement", "JobLevel", "JobSatisfaction", "PerformanceRating", "RelationshipSatisfaction", "WorkLifeBalance"]
# features[categorical_columns] = features[categorical_columns].astype('category')
```

```python
# features encoding
encoded_features = pd.get_dummies(features, drop_first=True)
encoded_features.head(5)
```

```python
print(encoded_features.info())
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
def scale_features(X, scaling_type='minmax'):
    if scaling_type == 'standard':
        scaler = StandardScaler()
    elif scaling_type == 'minmax':
        scaler = MinMaxScaler()
    
    continuous_cols = encoded_features.columns[encoded_features.dtypes != 'bool']
    X[continuous_cols] = scaler.fit_transform(X[continuous_cols])
    return X
```

```python
scaled_features = scale_features(encoded_features, 'minmax')
scaled_features.head(5)
```

# Correlation Analysis

```python
# Convert to dataframe
labels_df = pd.DataFrame(encoded_labels_np, columns=["Attrition"])
```

```python
# Correlation matrix
scaled_df = pd.concat([scaled_features, labels_df], axis=1)
corr_matrix = scaled_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# correlation matrix visualization
plt.figure(figsize=(20, 18))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5, fmt='.2f', annot_kws={"size": 8})
plt.title('Correlation Matrix')
plt.show()
```

```python
# correlation with Attrition
correlation = corr_matrix['Attrition'].sort_values(ascending=False)
print(correlation[1:21])
```

```python
# Select the top 20 features with the highest absolute correlation
top_20_features_names = correlation.abs().sort_values(ascending=False).index[1:21]
```

```python
corr_attrition_top20 = corr_matrix.loc[['Attrition'], top_20_features_names]
plt.figure(figsize=(15, 1))
sns.heatmap(corr_attrition_top20, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation of Top 20 Features with Attrition', fontsize=16)
plt.show()
```

# Scatter plot

```python
attrition_yes = scaled_features[top_20_features_names].loc[labels_df["Attrition"] == 1]
attrition_no = scaled_features[top_20_features_names].loc[labels_df["Attrition"] == 0]

attrition_yes.shape, attrition_no.shape
```

```python
# 1d scatter plot for top 20 features (4 rows, 5 columns)
# above yes and below no
fig, axs = plt.subplots(4, 5, figsize=(20, 16)) 
fig.suptitle('1D Scatter Plots of Top 20 Features by Attrition Class', fontsize=16)
for i, feature in enumerate(top_20_features_names):
    row = i // 5
    col = i % 5

    axs[row, col].plot(attrition_no[feature], np.zeros_like(attrition_no[feature]), 'o', label='No', alpha=0.5)
    axs[row, col].plot(attrition_yes[feature], np.zeros_like(attrition_yes[feature]), 'o', label='Yes', alpha=0.5)
    
    axs[row, col].set_title(f'{feature}', fontsize=12)
    axs[row, col].set_yticks([])
    axs[row, col].legend()

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# above no and below yes
fig, axs = plt.subplots(4, 5, figsize=(20, 16)) 
fig.suptitle('1D Scatter Plots of Top 20 Features by Attrition Class', fontsize=16)
for i, feature in enumerate(top_20_features_names):
    row = i // 5
    col = i % 5

    axs[row, col].plot(attrition_yes[feature], np.zeros_like(attrition_yes[feature]), 'o', label='Yes', alpha=0.5)
    axs[row, col].plot(attrition_no[feature], np.zeros_like(attrition_no[feature]), 'o', label='No', alpha=0.5)
    
    axs[row, col].set_title(f'{feature}', fontsize=12)
    axs[row, col].set_yticks([])
    axs[row, col].legend()

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# 2d plot
plt.figure(figsize=(30, 24))
for i, feature in enumerate(top_20_features_names):
    plt.subplot(5, 4, i + 1)
    sns.stripplot(x='Attrition', y=feature, data=scaled_df, jitter=True)
    plt.title(f'{feature} vs Attrition')

plt.tight_layout()
plt.show()
```

```python
top_20_features = scaled_features[top_20_features_names]
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# with top 20 features
X_train, X_test, y_train, y_test = train_test_split(top_20_features, encoded_labels_np, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Logistic Regression classifier: {accuracy:.2f}")
```

```python
# with all features
X_train, X_test, y_train, y_test = train_test_split(scaled_features, encoded_labels_np, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Logistic Regression classifier: {accuracy:.2f}")
```
